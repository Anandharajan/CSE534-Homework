```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def aimd(x, y):
    '''Addictive Increase Multiplicative Decrese
    
       This is the real case and the fair case
    '''
    total = x + y
    points = [[x, y]]
    for i in range(0, 30):
        size = len(points)
        point = points[size-1]
        x = point[0]
        y = point[1]
        if x + y < 20:  # increase
            x += 1
            y += 1
        else:           # decrease
            x /= 2
            y /= 2
        points.append([x, y])
    return points
```

```python
array = np.transpose(aimd(8, 18))

plt.figure(figsize=(7,7))
plt.plot(array[0], array[1], alpha=0.5, marker = 'o')
x1, y1 = [0, 20], [0, 20]
x2, y2 = [0, 20], [20, 0]
plt.plot(x1, y1, x2, y2, alpha=0.5, color='0')
plt.title('Fair: Addictive Increase Multiplicative Decrese')
plt.show()
```
